```python
import pandas as pd
import os
import glob
import json

files = glob.glob("../data/raw/*.json")

df_list = []

for file in files:
    with open(file, "r", encoding="utf-8") as f:
        data = json.load(f)   # loads standard JSON array
        
    temp_df = pd.DataFrame(data)

    # extract year from filename
    year = os.path.basename(file).split(".")[0]
    temp_df["year"] = year

    df_list.append(temp_df)

# combine all files
df = pd.concat(df_list, ignore_index=True)

print("Final Shape:", df.shape)
df.head()
```

```python
print(df["category"].value_counts().head(20))
```

```python
finance_categories = [
    "BUSINESS",
    "POLITICS",
    "THE WORLDPOST",
    "IMPACT"
]

df_finance = df[df["category"].isin(finance_categories)].copy()

print("Filtered Shape:", df_finance.shape)
print(df_finance["category"].value_counts())
```

```python
def map_sector(text):
    text = str(text).lower()

    if any(word in text for word in ["bank", "finance", "loan", "interest rate"]):
        return "Banking"

    elif any(word in text for word in ["software", "ai", "cloud", "tech", "microsoft", "google", "infosys"]):
        return "IT"

    elif any(word in text for word in ["pharma", "drug", "hospital", "healthcare", "vaccine"]):
        return "Pharma"

    elif any(word in text for word in ["oil", "gas", "energy", "solar", "power"]):
        return "Energy"

    elif any(word in text for word in ["car", "tesla", "automobile", "vehicle", "ev"]):
        return "Automobile"

    else:
        return "Others"
```

```python
df_finance["text"] = (
    df_finance["headline"].astype(str)
    + " "
    + df_finance["short_description"].astype(str)
)
```

```python
df_finance["sector"] = df_finance["text"].apply(map_sector)
```

```python
print(df_finance["sector"].value_counts())
```

```python
df_finance.to_csv(
    "../data/processed/finance_sector_dataset.csv",
    index=False
)
```
